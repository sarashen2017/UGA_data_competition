```python
from datetime import datetime as dt
import numpy as np
from functools import reduce
import pandas as pd
```

```python
df_test = pd.read_csv('/Users/shiyuandeng/Desktop/UGA/DataScienceCompetition/UGA Data Science Competition 2021 - 3242021 - 146 PM/Simulated_Data_Test.csv')
df_train=pd.read_csv('/Users/shiyuandeng/Desktop/UGA/DataScienceCompetition/UGA Data Science Competition 2021 - 3242021 - 146 PM/Simulated_Data_Train.csv')
df_validation=pd.read_csv('/Users/shiyuandeng/Desktop/UGA/DataScienceCompetition/UGA Data Science Competition 2021 - 3242021 - 146 PM/Simulated_Data_Validation.csv')
```

```python
df_test1 = df_test.drop(['States','ind_acc_XYZ'], 1)
df_train1= df_train.drop(['States','ind_acc_XYZ'], 1)
df_validation1=df_train.drop(['States','ind_acc_XYZ'], 1)
```

```python
import impyute as impy   #indXYZ
def EM_impute(df):
    impute=impy.em(df)
    impute.columns=list(df)
    return impute
```

```python
data_em_test=EM_impute(df_test1)
data_em_train=EM_impute(df_train1)
data_em_validation=EM_impute(df_validation1)
```

```python
data_em_validation.describe()
```

```python
index1=df_test[['States','ind_acc_XYZ']]
index2=df_train[['States','ind_acc_XYZ']]
index3=df_validation[['States','ind_acc_XYZ']]
```

```python
Df_test=data_em_test.join(index1)
Df_train=data_em_train.join(index2)
Df_validation=data_em_validation.join(index3)
```
