```python
## KNN imputation ###
```

```python


from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2)
```

```python
def imput(dataset):
    data = pd.DataFrame(imputer.fit_transform(dataset.iloc[:,0:19]))
    data['States']=dataset['States']
    data['Default_ind']=dataset['Default_ind']
    data.columns=list(dataset)
    return (data)
test=imput(test)
data=imput(data)
vali=imput(validation)
```

```python
test.to_csv('KNN_impute_test.csv')
data.to_csv('KNN_impute.csv')
vali.to_csv('KNN_impute_validation.csv')
```
